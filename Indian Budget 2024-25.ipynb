```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Downloads/47B23710.xlsx'
```

```python
# Load the specific sheet by name
sheet_name = 'Budget at a Glance'
df = pd.read_excel(file_path, sheet_name=sheet_name)

df.set_index('Category', inplace=True)

# Plot the bar graph
df.plot(kind='bar', figsize=(14, 8))
plt.title('Budget at a Glance 2024-25 (Rs crore)')
plt.xlabel('Category')
plt.ylabel('Amount (Rs crore)')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
sheet_name = 'Receipts Breakup'
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Set the Category as the index
df.set_index('Category', inplace=True)

# Plot the bar graph
df.plot(kind='bar', figsize=(14, 8))
plt.title('Budget at a Glance 2024-25 (Rs crore)')
plt.xlabel('Category')
plt.ylabel('Amount (Rs crore)')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
sheet_name = 'Expenditure Breakup'
# Read the data from the Excel sheet
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Set the Category as the index
df.set_index('Category', inplace=True)

# Plot a line graph
df[['Actuals 2022-23', 'Budgeted 2023-24', 'Revised 2023-24', 'Budgeted 2024-25']].plot(kind='line', figsize=(14, 8), marker='o')
plt.title('Central Government Expenditure Over Years (Rs crore)')
plt.xlabel('Category')
plt.ylabel('Amount (Rs crore)')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
# Read the data from the Excel sheet
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Set the Category as the index
df.set_index('Category', inplace=True)

# Plot a heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(df[['Actuals 2022-23', 'Budgeted 2023-24', 'Revised 2023-24', 'Budgeted 2024-25']], annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Heatmap of Central Government Expenditure (Rs crore)')
plt.xlabel('Year')
plt.ylabel('Category')
plt.show()
```
